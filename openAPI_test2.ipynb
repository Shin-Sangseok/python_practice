```python
import requests

header_data = {
    'accept': 'application/json',
    'Authorization': '',
}

param_data = (
    ('page', '1'),
    ('perPage', '100'),
    ('serviceKey', ''),
)

url = ''
## decode key가 적용되었음.
## api마다 연결이 되지 않는다고 한다면, encode, decode를 바꾸어 테스트해볼 것.
response = requests.get(url, headers=header_data, params=param_data)
response
```

```python
json_data = response.json()
json_data
```

```python
import pickle #파일을 저장할 목적으로.. 서버에 파일을 올려놓고 싶을때 자주 사용(byte단위)
```

```python
with open('covid-gokr.json', 'wb') as fp: #자동으로 close되는 기능이 있음
    pickle.dump(json_data, fp)
```
