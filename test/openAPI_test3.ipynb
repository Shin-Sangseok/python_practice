```python
# Python3 샘플 코드 #


import requests

url = '' #충북 관광 
params ={'serviceKey' : '',
         'currentPage' : '1', 
         'perPage' : '100',
         'SIGUN' : '', 
         'EMD_NM' : '', 
         'TRRSRT_CNT' : '',
         'TRRSRT_NM' : '',
         'TRRSRT_TURIST_CNT' : '',
         'EMD_TURIST_CNT' : '', 
         'EMD_LDGS_CNT' : '',
         'LDGS_TURIST_CNT' : '' }

response = requests.get(url, params=params)
print(response.content)
```

```python
json_data = response.json()
json_data
```

```python
data_values[0]['SIGUN']
data_values[0]['EMD_NM']
data_values[0]['TRRSRT_CNT']
data_values[0]['TRRSRT_NM']
data_values[0]['TRRSRT_TURIST_CNT']
data_values[0]['EMD_TURIST_CNT']
data_values[0]['EMD_LDGS_CNT']
data_values[0]['LDGS_TURIST_CNT']
```

```python
total_row = []#0~99
for i in range(100):
    sigun = data_values[i]['SIGUN']
    emd = data_values[i]['EMD_NM']
    trrst_cnt = data_values[i]['TRRSRT_CNT']
    trrst_name = data_values[i]['TRRSRT_NM']
    trrst_tpeople_cnt = data_values[i]['TRRSRT_TURIST_CNT']
    emd_tpeople_cnt = data_values[i]['EMD_TURIST_CNT']
    emd_ldgs_cnt = data_values[i]['EMD_LDGS_CNT']
    ldgs_tpeople_cnt = data_values[i]['LDGS_TURIST_CNT']
    one_row = []
    one_row.append(sigun)
    one_row.append(emd)
    one_row.append(trrst_name)
    one_row.append(trrst_cnt)
    one_row.append(trrst_tpeople_cnt)
    one_row.append(emd_tpeople_cnt)
    one_row.append(emd_ldgs_cnt)
    one_row.append(ldgs_tpeople_cnt)
    print(one_row)
    total_row.append(one_row)
print('total row count', len(total_row))
```

```python
son_sub_tourist = []
def fedit():
    with open('cb_tour.json', 'wb') as fp:
        pickle.dump(total_row, fp)
    with open('cb_tour.json','rb') as fp:
         json_sub_tourist = pickle.load(fp)
    json_sub_tourist
    print(json_sub_tourist)
    return json_sub_tourist
```
