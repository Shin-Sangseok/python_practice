```python
import pandas as pd 
```

```python
df = pd.read_csv('../csv-data/folium-data.csv', encoding='utf-8')
df.head()
```

```python
del df['no']
```

```python
df.head()
```

```python
df.set_index('자치구', inplace=True)
df.head()
```

```python
map = folium.Map(location=[37.57,126.98],
                zoom_start=10
                )
map.choropleth(geo_data = geo_str,
               data = df['대기업'], 
               columns = [df.index, df['대기업']], 
               fill_color = 'PuRd', 
               key_on = 'feature.id'
              )
map
```

```python
map = folium.Map(location=[37.57, 126.98],
                zoom_start=10
                )
map.choropleth(geo_data = geo_str,
               data = df['중기업'], 
               columns = [df.index, df['중기업']],
               fill_color = 'PuRd', 
               key_on = 'feature.id'
              )
map
```

```python
map = folium.Map(location=[37.57, 126.98],
                zoom_start=10 
                )
map.choropleth(geo_data = geo_str,
              data = df['소기업'], 
              columns = [df.index, df['소기업']], 
               fill_color = 'PuRd', 
               key_on = 'feature.id'
              ) 
map
```

```python
# 중기업, 소기업
map = folium.Map(location=[37.57, 126.98], 
                zoom_start=10
                )
map.choropleth(geo_data=geo_str,
                      data= df['대기업'],
                      columns=[df.index, df['대기업']],
                      fill_color='PuRd',
                      key_on='feature.id'
              )


for n in df.index:
    folium.CircleMarker([df['lat'][n], df['lng'][n]], 
                        radius = df['대기업'][n]*0.1, 
                        color='#3186cc', fill_color='#3186cc', fill=True).add_to(map)
map
```
