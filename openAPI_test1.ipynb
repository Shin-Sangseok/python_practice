```python
import requests

url = ''
params ={'serviceKey' : '', 
         'pageNo' : '1', 
         'numOfRows' : '3', 
         'type' : 'json', 
         'YYYY' : '2020', 'MM' : '12' }

response = requests.get(url, params=params)
print(response.content)
```

```python
import requests

header_data = {
    'accept': 'application/json',
    'Authorization': '',
}

param_data = (
    ('page', '1'),
    ('perPage', '2'),
    ('serviceKey', ''),
)

url = ''
## decode key가 적용되었음.
## api마다 연결이 되지 않는다고 한다면, encode, decode를 바꾸어 테스트해볼 것.
response = requests.get(url, headers=header_data, params=param_data)
response
```

```python
# Python3 샘플 코드 #


import requests

url = ''
params ={'serviceKey' : '',
         'returnType' : 'JSON', 
         'numOfRows' : '10',
         'cond[country_nm::EQ]' : '가나',
         'cond[country_iso_alp2::EQ]' : 'GH', 
         'pageNo' : '1' }

response = requests.get(url, params=params)
print(response.content)
```
